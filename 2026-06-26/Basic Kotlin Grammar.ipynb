## 함수와 데이터 타입
- 반환문이 없는 경우: Unit 사용
- Unit 반환문 생략 가능

```kotlin
fun add(a : Int, b : Int): Int
{
    return a + b
}

println("2 + 3 = ${add(2, 3)}")

fun showMessage(mesg : String) : Any
{
    println(mesg)
    return 0
}

showMessage("안녕하세요")
```

```kotlin
var data7 : Any = 10
println("초기 값: $data7")

data7 = 20.3
println("Double로 변환: $data7")

data7 = '한'
println("Char로 변환: $data7")
```

```kotlin
var data8 : Int = 40
println("초기 Int: $data8")

var data9 : Double = 21.3

data8 = data9.toInt()
println("Double로 변환: $data8")
```

### 함수의 기본 값 정의
- 함수 인수에 값을 지정하면, 가변 인수를 전달할 수 있음

```kotlin
import kotlin.math.pow

fun power(base : Double, exp : Int = 2) : Double
{
    return base.pow(exp)
}

println("2^3 = ${power(2.0, 3)}")
println("3^2 = ${power(3.0)}")
```

## if 제어문
- 일반 프로그래밍 언어와 동일
- 삼항 연산자처럼 if 조건식에 따라 값을 지정 가능

```kotlin
import kotlin.random.Random

val score = Random.nextInt(0, 101) // 0 - 100까지
val cut_off = if (score >= 60)
    {
        println("점수를 비교해서, String 합격 여부를 반환.")
        "Dummy"
        "Pass"
    }
    else "Fail"

println("너의 점수는 ${score}이라서 ${cut_off}야.")
```

## when 제어문 사용
- when 제어문은 기존의 switch 문보다 강력함
- 일반적인 switch는 if 동등절에 해당하나, when 제어문은 범위 및 다양한 조건을 포함할 수 있음

```kotlin
// print("입력 스트링: ")

// val input : String = readln()
val data : Any = "Test"

when(data)
{
    is String -> println("입력이 String임.")
    20, 30 -> println("입력 값이 20 또는 30임")
    in 1..10 -> println("입력 값이 1이상 10이하")
    else -> println("입력 값이 예상 범위를 벗어남")
}
```

## 반복문을 통해서 원주율 계산하기
- pi/4 = 1 - 1/3 + 1/5 - 1/7 + ...

```kotlin
fun series() {
    for (t in listOf(100, 1000, 5000)) {
        var total : Double = 0.0
        var sign = 1
        for (i in 0 until t) {
            total += sign / (2.0 * i + 1)
            sign = -sign
        }
        println("${t}항: ${total * 4}")
    }
}
series()
```

```kotlin
fun series(n : Int) : Double
{
    var x : Double = 0.0
    for (i in 1..n)
    {
        val denom = 2 * i - 1
        val sign = if (i % 2 == 1) 1 else -1
        x += sign / denom.toDouble()
    }
    return x * 4
}

println("Series(100) = ${series(100)}")
println("Series(1000) = ${series(1000)}")
println("Series(5000) = ${series(5000)}")
```
